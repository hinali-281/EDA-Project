```python
from sklearn.linear_model import LinearRegression
# Create a copy to work with
df_imputed=df.copy()
```

```python
# Use rows where 'Income' is not missing
df_complete = df_imputed[df_imputed['Income'].notnull()]
df_missing = df_imputed[df_imputed['Income'].isnull()]
```

```python
# Choose predictors
predictors = ['Age', 'Credit_Score', 'Credit_Utilization', 'Loan_Balance', 'Debt_to_Income_Ratio']
df_complete = df_complete.dropna(subset=predictors)
```

```python
# Prepare training data
X_train = df_complete[predictors]
y_train = df_complete['Income']
```

```python
# Fit linear regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict missing income
X_missing = df_missing[predictors].dropna()
predicted_income = model.predict(X_missing)

# Get residual standard deviation (to simulate noise)
residuals = y_train - model.predict(X_train)
residual_std = residuals.std()

# Monte Carlo simulation: Add random noise
np.random.seed(42)
simulated_income = predicted_income + np.random.normal(0, residual_std, size=len(predicted_income))

# Fill back into dataset
df_imputed.loc[X_missing.index, 'Income'] = simulated_income

# Show a few imputed results
print(df_imputed.loc[X_missing.index, ['Age', 'Credit_Score', 'Income']].head())
```

```python
#for Loan_Balance
# Work with df_mc (which already has imputed Income)
loan_known = df_imputed['Loan_Balance'].dropna()
loan_missing_idx = df_imputed[df_imputed['Loan_Balance'].isnull()].index

# Estimate mean and std from known Loan_Balance
mean_loan = loan_known.mean()
std_loan = loan_known.std()

# Simulate missing values from a normal distribution
np.random.seed(42)
simulated_loans = np.random.normal(loc=mean_loan, scale=std_loan, size=len(loan_missing_idx))

# Ensure no negative values
simulated_loans = np.clip(simulated_loans, a_min=0, a_max=None)

# Fill missing Loan_Balance values in df_mc
df_imputed.loc[loan_missing_idx, 'Loan_Balance'] = simulated_loans

# Preview results
df_imputed.loc[loan_missing_idx, ['Age', 'Credit_Score', 'Loan_Balance']].head()
```

```python
#for Credit_Score
# Identify missing credit score indices
credit_missing_idx = df_imputed[df_imputed['Credit_Score'].isnull()].index

# Extract known values
known_credit_scores = df_imputed['Credit_Score'].dropna().values

# Set random seed for reproducibility
np.random.seed(42)

# Sample with replacement (bootstrapping)
simulated_credit_scores = np.random.choice(known_credit_scores, size=len(credit_missing_idx), replace=True)

# Fill in the missing values
df_imputed.loc[credit_missing_idx, 'Credit_Score'] = simulated_credit_scores

# Preview the filled values
df_imputed.loc[credit_missing_idx, ['Age', 'Credit_Score', 'Income']].head()
```
